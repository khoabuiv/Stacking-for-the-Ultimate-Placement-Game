```python
import pygame
import math

pygame.init()
#Font
Font = pygame.font.SysFont('Arial',35)
```

```python
#Colors
black = (0, 0, 0)
white = (255, 255, 255)
red = (255, 0, 0)
blue = (0,0,255)
grey = (220,220,220)

window_size = (700,700)

position = []
blue_possible = [] #Auxilary list to represent player 2 move on the small board
red_possible = [] #Auxilary list to represent player 1 move on the small board
Win_con_op = [] #Auxilary list for the big board
Big_board_position = [] #Position on the big board. 
Big_board_possible_red = []
Big_board_possible_blue = []
ref_N = []



board_size = 2
next_game = board_size + 1


for row in range(board_size):
    position.append([])
    blue_possible.append([])
    red_possible.append([])
    Win_con_op.append(0)
    Big_board_position.append(0)
    Big_board_possible_red.append('Y')
    Big_board_possible_blue.append('Y')
    ref_N.append('N')
    for column in range(board_size):
        position[row].append(0)
        blue_possible[row].append('Y')
        red_possible[row].append('Y')
        

#All the surfaces we will use
static_img = pygame.Surface(window_size,pygame.SRCALPHA) #Biggest circles
big_boards_static = pygame.Surface(window_size,pygame.SRCALPHA) #The outside of the board
big_boards_active = pygame.Surface(window_size,pygame.SRCALPHA)
small_boards_static = pygame.Surface(window_size,pygame.SRCALPHA)
small_boards_active = pygame.Surface(window_size,pygame.SRCALPHA)
blinking = pygame.Surface(window_size,pygame.SRCALPHA)

####

#Start Drawing Board
Main_center = (window_size[0] // 2, window_size[1] // 2)
big_radius = 260
big_board_radius = 87
big_board_ref_radius = big_board_radius-40
small_board_radius = 17
pygame.draw.circle(static_img, black, Main_center, big_radius, width = 5)
degree = math.radians(360 // board_size)
big_board_center_list = [] #We need this list to change game state later
small_board_center_list = []

for i in range(board_size):
    big_board_center = (Main_center[0] - big_radius*math.sin(degree*i), Main_center[1] - big_radius*math.cos(degree*i))
    big_board_center_list.append(big_board_center)
    small_board_center_list.append([])
    pygame.draw.circle(big_boards_active, black, big_board_center_list[i], big_board_radius, width = 5)
    pygame.draw.circle(big_boards_static, white, big_board_center_list[i], big_board_radius-5)
    pygame.draw.circle(small_boards_static, black, big_board_center_list[i],big_board_ref_radius, width = 3)
    pygame.draw.circle(small_boards_active, black, big_board_center_list[i],small_board_radius, width = 3)
    for j in range(board_size):
        small_board_center = (big_board_center_list[i][0] - (big_board_ref_radius)*math.sin(degree*j), 
                                       big_board_center_list[i][1] - (big_board_ref_radius)*math.cos(degree*j))
        small_board_center_list[i].append(small_board_center)
        pygame.draw.circle(small_boards_active, black, small_board_center_list[i][j], small_board_radius, width = 3)
        pygame.draw.circle(small_boards_static, white, small_board_center_list[i][j], small_board_radius-3)
#End Drawing Board

#Winner Screen
P1_Win = Font.render('Player 1 Wins!' , True , red)
P1_Win_size = Font.size('Player 1 Wins!')
P2_Win = Font.render('Player 2 Wins' , True , blue)
P2_Win_size = Font.size('Player 2 Wins!')


screen = pygame.display.set_mode(window_size)
pygame.display.set_caption("Snort on a Cycle")
    
done = False
clock = pygame.time.Clock()



k = 0
Winner = 0


alpha = 1
change = True
done = False
blink_spd = 0.6
pos = (0,0)

while not done:
    for event in pygame.event.get(): 
        if event.type == pygame.QUIT: 
            done = True 
        elif event.type == pygame.MOUSEBUTTONDOWN:
            pos = pygame.mouse.get_pos()

    #Blinking effect      
    if change == True:
        alpha = alpha + blink_spd
        if alpha >= 175:
            change = False
    elif change == False:
        alpha = alpha - blink_spd
        if alpha <= 30:
            change = True
       
    #Winning Condition
    if ((all(i == 'N' for i in Big_board_possible_red) == True and Winner == 0 and k % 2 == 0)
        or ((all(i == ref_N for i in red_possible)) and k % 2 == 0 and Winner == 0 
            and all(i != 1 for i in Win_con_op) == True)):
        Winner = 2
    elif ((all(i == 'N' for i in Big_board_possible_blue) == True and Winner == 0 and k % 2 == 1)
        or ((all(i == ref_N for i in blue_possible)) and k % 2 == 1 and Winner == 0
            and all(i != 2 for i in Win_con_op) == True)):
        Winner = 1
                
    #Move set instructions
    for i in range(board_size):    
        # Draw for the big board
        if Big_board_position[i] == 1:
            pygame.draw.circle(big_boards_active, red, big_board_center_list[i], big_board_radius, width = 5)
            pygame.draw.circle(big_boards_active, red, big_board_center_list[i], big_board_ref_radius, width = 8)
            pygame.draw.circle(small_boards_active, red, big_board_center_list[i],small_board_radius, width = 0)
            pygame.draw.circle(small_boards_static, white, big_board_center_list[i],big_board_ref_radius, width = 3)
        elif Big_board_position[i] == 2:
            pygame.draw.circle(big_boards_active, blue, big_board_center_list[i], big_board_radius, width = 5)
            pygame.draw.circle(big_boards_active, blue, big_board_center_list[i], big_board_ref_radius, width = 8)
            pygame.draw.circle(small_boards_active, blue, big_board_center_list[i],small_board_radius, width = 0)
            pygame.draw.circle(small_boards_static, white, big_board_center_list[i],big_board_ref_radius, width = 3)
        #Draw the blinking effect for big board
        if (Win_con_op[i] == 1 
              and Big_board_possible_red[i] == 'Y' 
              and Big_board_position[i] == 0 and k % 2 == 0):
            pygame.draw.circle(blinking, red,big_board_center_list[i], big_board_radius + 3, width = 3)
            pygame.draw.circle(blinking, red, big_board_center_list[i],small_board_radius - 3, width = 3)
        elif (Win_con_op[i] == 2 
              and Big_board_possible_blue[i] == 'Y' 
              and Big_board_position[i] == 0 and k % 2 == 1):
            pygame.draw.circle(blinking, blue,big_board_center_list[i], big_board_radius + 3, width = 3)
            pygame.draw.circle(blinking, blue, big_board_center_list[i],small_board_radius - 3, width = 3)
        else:
            pygame.draw.circle(blinking,white,big_board_center_list[i], big_board_radius + 3, width = 3)
            pygame.draw.circle(blinking, white, big_board_center_list[i],small_board_radius - 3, width = 3)
        
        #Check the board state
        if (all(j == 'N' for j in red_possible[i]) == True and k % 2 == 0):
            if next_game == i:
                next_game = board_size + 1 #If there is no possible move on a board i, then the player free to move anywhere
            if (Win_con_op[i] == 0):
                Win_con_op[i] = 2 # small boards winning conditions checks
        if (all(j == 'N' for j in blue_possible[i]) == True and k % 2 == 1):
            if next_game == i:
                next_game = board_size + 1
            if (Win_con_op[i] == 0):
                Win_con_op[i] = 1 # small boards winning conditions checks
                    
        for j in range(board_size):
            #Big Board Instruction
            if ((((big_board_center_list[i][0] - small_board_radius) < pos[0]
                and pos[0] < (big_board_center_list[i][0] + small_board_radius))
                and ((big_board_center_list[i][1] - small_board_radius) < pos[1]
                and pos[1] < (big_board_center_list[i][1] + small_board_radius)))
                 ):
                if Win_con_op[i] == 1 and k % 2 == 0:
                    if Big_board_position[(i+1) % len(position)] == 2 or Big_board_position[(i-1) % len(position)] == 2:
                        Big_board_position[i] = 0
                    else:
                        Big_board_position[i] = 1
                        next_game = board_size + 1
                        k = k + 1
                        Big_board_possible_red[(i) % len(position)] = 'N'
                        Big_board_possible_blue[(i) % len(position)] = 'N'
                        Big_board_possible_blue[(i+1) % len(position)] = 'N'
                        Big_board_possible_blue[(i-1) % len(position)] = 'N'
                        Win_con_op[i] = 0
                elif Win_con_op[i] == 2 and k % 2 == 1: 
                    if Big_board_position[(i+1) % len(position)] == 1 or Big_board_position[(i-1) % len(position)] == 1:
                        Big_board_position[i] = 0
                    else:
                        Big_board_position[i] = 2
                        next_game = board_size + 1
                        k = k + 1
                        Big_board_possible_blue[(i) % len(position)] = 'N'
                        Big_board_possible_red[(i) % len(position)] = 'N'
                        Big_board_possible_red[(i+1) % len(position)] = 'N'
                        Big_board_possible_red[(i-1) % len(position)] = 'N'
                        Win_con_op[i] = 0
            #Small Board Instructions
            elif ((small_board_center_list[i][j][0] - small_board_radius) < pos[0] 
                and pos[0] < (small_board_center_list[i][j][0] + small_board_radius) 
                and (small_board_center_list[i][j][1] - small_board_radius) < pos[1] 
                and pos[1] < (small_board_center_list[i][j][1] + small_board_radius) 
                ):
                if (k % 2 == 0 
                and position[i][j] == 0 
                and (next_game == i or next_game == board_size + 1)):
                    if position[i][(j-1) % len(position)] == 2 or position[i][(j+1) % len(position)] == 2:
                        position[i][j] = 0
                    else:
                        position[i][j] = 1
                        next_game = j
                        red_possible[i][(j) % len(position)] = 'N'
                        blue_possible[i][(j) % len(position)] = 'N'
                        blue_possible[i][(j-1) % len(position)] = 'N'
                        blue_possible[i][(j+1) % len(position)] = 'N'
                        k = k + 1
                elif (k % 2 == 1 
                and position[i][j] == 0 
                and (next_game == i or next_game == board_size + 1)):
                    if position[i][(j-1) % len(position)] == 1 or position[i][(j+1) % len(position)] == 1:
                        position[i][j] = 0
                    else:
                        position[i][j] = 2
                        next_game = j
                        blue_possible[i][(j) % len(position)] = 'N'
                        red_possible[i][(j) % len(position)] = 'N'
                        red_possible[i][(j-1) % len(position)] = 'N'
                        red_possible[i][(j+1) % len(position)] = 'N'
                        k = k + 1
    #Draw on the small board
            if position[i][j] == 1:
                pygame.draw.circle(small_boards_active,red,small_board_center_list[i][j],small_board_radius-3,width=0)
            elif position[i][j] == 2:
                pygame.draw.circle(small_boards_active,blue,small_board_center_list[i][j],small_board_radius-3,width=0) 
    #Blinking Effect on the small board
            if (red_possible[i][j] == 'Y' 
                  and position[i][j] == 0
                  and k % 2 == 0 
                  and (next_game == board_size + 1 or next_game == i)):
                pygame.draw.circle(blinking,red,small_board_center_list[i][j],small_board_radius-3,width=3)
            elif (blue_possible[i][j] == 'Y' 
                  and position[i][j] == 0
                  and k % 2 == 1 
                  and (next_game == board_size + 1 or next_game == i)):
                pygame.draw.circle(blinking,blue,small_board_center_list[i][j],small_board_radius-3,width=3)
            else:
                pygame.draw.circle(blinking,white,small_board_center_list[i][j],small_board_radius-3,width=3)
                                                       

    blinking.set_alpha(alpha)        
    pos = (0,0)   
            
    screen.fill(white) #Background color 
    screen.blit(static_img,(0,0))
    screen.blit(big_boards_static,(0,0))
    screen.blit(big_boards_active,(0,0))
    screen.blit(small_boards_static,(0,0))
    screen.blit(blinking,(0,0))
    screen.blit(small_boards_active,(0,0))
    if Winner == 1:
        screen.blit(P1_Win,(0,0))
        pygame.draw.circle(static_img, red, Main_center, big_radius, width = 5)
    elif Winner == 2:
        screen.blit(P2_Win,(0,0))
        pygame.draw.circle(static_img, blue, Main_center, big_radius, width = 5)
        
        
    
    
    
    pygame.display.flip()
```

```python
pygame.quit()
```
