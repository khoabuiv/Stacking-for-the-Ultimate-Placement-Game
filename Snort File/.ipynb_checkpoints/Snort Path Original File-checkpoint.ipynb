```python
import pygame
#Color for drawing
black = (0, 0, 0)
white = (255, 255, 255)
red = (255, 0, 0)
blue = (0,0,255)


position = []
Check = []
Win_con_op = []

big_cell = 3 #number of big cells that we want. Works well from 2-6, after that the ratio looks weird.
cell_r = 75 #cell radius
distance_between_cell = 25 #Distance between cells
distance_between_radius = cell_r * 2 + distance_between_cell
next_game = big_cell+1



for row in range(big_cell):
    position.append([])
    Check.append([])
    Win_con_op.append(0)
    for column in range(big_cell):
        position[row].append(0)
        Check[row].append(0)

    
pygame.init()


window_size = [(big_cell * cell_r * 2) + ((big_cell) * distance_between_cell), cell_r*2]
width = 330 - (window_size[0] // 2)
height = 330 - (window_size[1] // 2)

size = 660,660
screen = pygame.display.set_mode(size)
pygame.display.set_caption("Snort")


board_img = pygame.Surface(window_size,pygame.SRCALPHA)
board_img.fill((255,255,255,0))

#Start drawing the board#
Big_circle_center = []
Small_circle_center = []
for circle in range(big_cell): #Outside game
    center_big = ((cell_r+distance_between_cell/2)+(circle*distance_between_radius),cell_r)
    Big_circle_center.append(center_big)
    pygame.draw.circle(board_img,black,center_big, cell_r,width=2)
    start = (cell_r+distance_between_cell/2)+(circle*distance_between_radius) - cell_r 
    end = (cell_r+distance_between_cell/2)+(circle*distance_between_radius) + cell_r
    interval = (end-start) / big_cell
    small_r = (cell_r / big_cell) / 1.5
    small_distance_between_radius = small_r + interval 
    Small_circle_center.append([])
    pygame.draw.line(board_img,black,(interval + (circle * distance_between_radius),cell_r),((big_cell-((big_cell-1)*0.5))*interval + big_cell*small_r + (circle * distance_between_radius),cell_r))
    for circle_small in range(big_cell): #Inside game board
        center_small = ((small_r + interval/2) + (circle_small * interval) + (circle * distance_between_radius), cell_r)
        Small_circle_center[circle].append(center_small)
        pygame.draw.circle(board_img,black,center_small,small_r,width=3)
        pygame.draw.circle(board_img,white,center_small,small_r-3,width=0)

for circle in range(big_cell-1): #Connect line for outside game
    pygame.draw.line(board_img,black,(2*cell_r + circle*distance_between_radius + distance_between_cell/2,cell_r), (2*cell_r + circle*distance_between_radius + 1.5*distance_between_cell,cell_r))

#End drawing the board#


done = False
clock = pygame.time.Clock()
k = 0

alpha = 1
change = True
blink_spd = 0.15
pos = (0,0)


while not done:      
    #Mouseclick event
    for event in pygame.event.get(): 
        if event.type == pygame.QUIT: 
            done = True 
        elif event.type == pygame.MOUSEBUTTONDOWN:
            pos = pygame.mouse.get_pos()
    
    #Move set instructions
    for i in range(len(Small_circle_center)):
        for j in range(len(Small_circle_center[i])):
            if (width + Small_circle_center[i][j][0] - small_r) < pos[0] < (width + Small_circle_center[i][j][0] + small_r) and (height + Small_circle_center[i][j][1] - small_r) < pos[1] < (height + Small_circle_center[i][j][1] + small_r) and k % 2 == 0:
            # Player 1 select a circle
                if next_game == i or next_game == big_cell+1:
                    if j == 0: #Left edge
                        if position[i][j+1] == 2: 
                            position[i][j] = 0 
                        else:
                            position[i][j] = 1
                            next_game = j
                            k = k + 1
                    elif j == (len(Small_circle_center[i]) - 1): #Right edge
                        if position[i][j-1] == 2:
                            position[i][j] = 0
                        else:
                            position[i][j] = 1
                            next_game = j
                            k = k + 1
                    else: #Normal
                        if position[i][j-1] == 2 or position[i][j+1] == 2:
                            position[i][j] = 0
                        else:
                            position[i][j] = 1
                            next_game = j
                            k = k + 1
            elif (width + Small_circle_center[i][j][0] - small_r) < pos[0] < (width + Small_circle_center[i][j][0] + small_r) and (height + Small_circle_center[i][j][1] - small_r) < pos[1] < (height + Small_circle_center[i][j][1] + small_r) and k % 2 == 1:
            # Player 2 select a circle
                if next_game == i or next_game == big_cell+1:
                    if j == 0: #Left edge
                        if position[i][j+1] == 1: 
                            position[i][j] = 0 
                        else:
                            position[i][j] = 2
                            next_game = j
                            k = k + 1
                    elif j == (len(Small_circle_center[i]) - 1): #Right edge
                        if position[i][j-1] == 1:
                            position[i][j] = 0
                        else:
                            position[i][j] = 2
                            next_game = j
                            k = k + 1
                    else: #Normal
                        if position[i][j-1] == 1 or position[i][j+1] == 1:
                            position[i][j] = 0
                        else:
                            position[i][j] = 2
                            next_game = j
                            k = k + 1  
                    
    #Inner boards winning conditions checks
    for i in range(len(position)): #Copy from the position list to the check list. I use position list to display, and the check list as the back
        for j in range(len(position[i])):
            if position[i][j] != 0:
                Check[i][j] = position[i][j]
        
    for i in range(len(position)): #The idea here is that if a player selects a vertex, then he/she also claims to adjaccent vertices on the check list
        for j in range(len(position[i])):
            if position[i][j] == 0: #Only check the empty vertex   
                if j == 0: #If left side edge, then only check one case
                    if position[i][1] == 1 and k % 2 == 1:
                        Check[i][0] = 1
                    elif position[i][1] == 2 and k % 2 == 0:
                        Check[i][0] = 2
                elif j == (len(position[i])-1):
                    if position[i][j-1] == 1 and k % 2 == 1:
                        Check[i][j] = 1
                    elif position[i][j-1] == 2 and k % 2 == 0:
                        Check[i][j] = 2
                else:
                    for j in range(1,len(position[i])-1):
                        if (position[i][j-1] == 1 or position[i][j+1] == 1) and k % 2 == 1:
                            Check[i][j] = 1
                        elif (position[i][j-1] == 2 or position[i][j+1] == 2) and k % 2 == 0:
                            Check[i][j] = 2

    for i in range(len(Check)):
        if all(j != 0 for j in Check[i]):
            if Check[i].count(2) > Check[i].count(1):
                Win_con_op[i] = 2
            elif Check[i].count(1) > Check[i].count(2):
                Win_con_op[i] = 1
    
    #Draw on the board
    for row in range(big_cell): #Draw the cells
        for column in range(big_cell):
            Small_circle_center[row][column]
            if position[row][column] == 1:
                pygame.draw.circle(board_img,red,Small_circle_center[row][column],small_r-2,width=0)
            elif position[row][column] == 2:
                pygame.draw.circle(board_img,blue,Small_circle_center[row][column],small_r-2,width=0)
            
    pos = (0,0)    

    
    screen.fill((255,255,255))
    screen.blit(board_img,(330 - (window_size[0] // 2),330 - (window_size[1] // 2)))
    
    
    pygame.display.flip()
    
print(k)
print(position)
print(Win_con_op)
print(Check)
    
    
pygame.quit()
```
